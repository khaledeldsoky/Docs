### awk

#### Basic Syntax

```python
# Print each line of the file
awk '{print $0}' file.txt
# Print lines that match the pattern
awk'/pattern/ {print $0}' file.txt
# Print the first column of each line
awk '{print $1}' file.txt
# Print the first and third columns of each Line
awk '{print $1, $3}' file.txt
# Add a header and footer
awk 'BEGIN {print "Header"} {print $0} END {print "Footer"}' file.txt
```

#### Field and Record Separators

```python
# Use ":" as the field separator, print the first field
awk -F":" '{print $1}' /etc/passwd
# Set field separator ":" in the BEGIN block
awk 'BEGIN {FS=":"} {print $1}'/etc/passwd
# Use an empty line as the record separator
awk 'BEGIN (RS="\n\n"} {print $0}' file.txt
```

#### Pattern Matching

```python
# Print lines where the third field matches apattern
awk '$3/pattern/ {print $0}' file.txt
# Print lines where the third field does not match a pattern
awk '$3!/pattern/ {print $0}' file.txt
```

#### Arithmetic Operations

```python
# Add the first and second fields of each line
awk '{print $1 $2}' file.txt
#Sum the values of the first field across all Lines
awk '{sum += $1} END {print sum}' file.txt
# Calculate the average of the first three fields
awk '{avg ($1 $2 $3)/3; print avg}' file.txt
```

#### Working with Multiple Files

```python
# Print a message whenstarting a new file
awk 'FNR==1 {print "Processing" FILENAME}' file1.txt file2.txt
# Separate calculations for each file
awk 'FNR==NR {sum1 += $1; next} {sum2 += $1] END {print sum1, sum2}' file1.txt file2.txt
```

### base 64

```python
Encode a String to Base64

echo -n "my-secret-key" | base64

✅ Encode a File to Base64

base64 config.json

✅ If you want to save the output to another file, use:

base64 config.json > config.b64

✅ Decode Base64 Back to Normal Text

echo "bXktc2VjcmV0LWtleQ==" | base64 --decode

✅ For a Base64-encoded file

base64 --decode config.b64 > config.json
```
