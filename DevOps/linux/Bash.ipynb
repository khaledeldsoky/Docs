## IF

```python
#!/bin/bash
if [ "$var" == "value" ]; then
    echo "Match"
else
    echo "No match"
fi
```

### 1 - Check if a file exists :

```python
#!/bin/bash
file="test.txt"
if [ -f "$file" ]; then
    echo "$file exists"
else
    echo "$file does not exist"
fi
```

###  2 - Compare numbers :

```python
#!/bin/bash
num=15
if [ $num -gt 10 ]; then
    echo "$num is greater than 10"
elif [ $num -eq 10 ]; then
    echo "$num equals 10"
else
    echo "$num is less than 10"
fi
```

###  3 - Check string  :

```python
#!/bin/bash
user="Alice"
if [ "$user" = "Alice" ]; then
    echo "Welcome, $user!"
else
    echo "Unknown user"
fi
```

### 4 - Check if a directory is empty :

```python
#!/bin/bash
dir="myfolder"
if [ -z "$(ls -A $dir)" ]; then
    echo "$dir is empty"
else
    echo "$dir is not empty"
fi
```

#### 5- List Files in Current Directory

```python

#! /bin/bash

filename=$1

if [ -f "$filename" ]; then
  echo "file is exist"
else
  echo "file not exist"
fi
```

## Loops

### For

```python
for i in 1 2 3; do
    echo $i
done
```

#### 1 - Iterate over a list of words

```python
#!/bin/bash
for fruit in apple banana orange; do
    echo "Fruit: $fruit"
done
```

#### 2 -  Loop over files in a directory

```python
#!/bin/bash
for file in *.txt; do
    echo "Found file: ${file}"
done
```

#### 3 - Iterate over a range

```python
#!/bin/bash
for i in {1..5}; do
    echo "Number: $i"
done
```

#### 4 -Loop with command output

```python
#!/bin/bash
for line in $(cat names.txt); do
    echo "Name: $line"
done
```

## While 

```python
count=1
while [ $count -le 3 ]; do
    echo $count
    ((count++))
done
```

#### 1 - Count from 1 to 5 

```python
#!/bin/bash
count=1
while [ $count -le 5 ]; do
    echo "Count: $count"
    ((count++))
done
```

####  2 - Read a file line by line

```python
#!/bin/bash
file="input.txt"
while IFS= read -r line; do
    echo "Line: $line"
done < "$file"
```

####  3 - Wait for a file to exist

```python
#!/bin/bash
file="data.txt"
while [ ! -f "$file" ]; do
    echo "Waiting for $file to appear..."
    sleep 2
done
echo "$file found!"
```

####  4 - Infinite loop with user input

```python
#!/bin/bash
while true; do
    echo "Enter 'quit' to exit: "
    read input
    if [ "$input" = "quit" ]; then
        break
    fi
    echo "You entered: $input"
done
```
