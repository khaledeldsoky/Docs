### Date Code

- Year Code

````python
-- Active: 1770201048896@@10.20.20.50@1521@WRMICTTST@CSBASE

SELECT  EXTRACT(YEAR FROM SYSDATE) - 10 + LEVEL - 1 AS year_val
FROM DUAL
CONNECT BY LEVEL <= 21;


SELECT year_val
FROM (
    SELECT  EXTRACT(YEAR FROM SYSDATE) - 10 + LEVEL - 1 AS year_val
    FROM DUAL
    CONNECT BY LEVEL <= 21
)
ORDER BY year_val DESC
````

- Month Code

````python
SELECT 
  y.year_val,
  month_num,
  month_name
FROM (
  SELECT  EXTRACT(YEAR FROM SYSDATE) - 10 + LEVEL - 1 AS year_val
  FROM DUAL
  CONNECT BY LEVEL <= 21
) y
CROSS JOIN (
SELECT LEVEL AS month_num, 
       TO_CHAR(ADD_MONTHS(DATE '2000-01-01', LEVEL-1), 'Month') AS month_name
FROM DUAL
CONNECT BY LEVEL <= 12
) m
-- WHERE  y.year_val = $P{Year}

SELECT 
  y.year_val,
  m.month_num,
  TO_CHAR(ADD_MONTHS(DATE '2000-01-01', m.month_num - 1), 'Month') AS month_name
FROM (
  SELECT  EXTRACT(YEAR FROM SYSDATE) - 10 + LEVEL - 1 AS year_val
  FROM DUAL
  CONNECT BY LEVEL <= 21
) y
CROSS JOIN (
    SELECT LEVEL AS month_num
    FROM DUAL
    CONNECT BY LEVEL <= 12
) m
-- WHERE  y.year_val = ${Year}
ORDER BY y.year_val DESC, m.month_num
````

- Quarter

````python

SELECT 
  y.year_val,
  month,
  Quarter
FROM (
  SELECT  EXTRACT(YEAR FROM SYSDATE) - 10 + LEVEL - 1 AS year_val
  FROM DUAL
  CONNECT BY LEVEL <= 21
) y
CROSS JOIN (
SELECT 
  decode(LEVEL,'1','1,2,3','2','4,5,6','3','7,8,9','4','10,11,12' , LEVEL) as month,
  LEVEL as Quarter
FROM DUAL
CONNECT BY LEVEL <= 4
) m
-- WHERE  y.year_val = $P{Year}
````
